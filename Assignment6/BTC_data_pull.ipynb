```python
import requests

#https request parameters 
symbol_id = 'BITSTAMP_SPOT_BTC_USD'
period_id = 'period_id=1MTH'
output_format = 'output_format=csv'
time_start = 'time_start=2011-09-01T00:00:00'
time_end = 'time_end=2018-02-01T00:00:00'
limit = 'limit=1200'

#concatenate parameters to url request string
url = 'https://rest.coinapi.io/v1/ohlcv/'+symbol_id+'/history?'+period_id+'&'+output_format+'&'+time_start+'&'+time_end+'&'+limit
print url

headers = {'X-CoinAPI-Key' : 'xxxxxxxx-xxxx-xxxx-xxxx-xxxxxxxxxxxx'} #insert user generated key
response = requests.get(url, headers=headers)
```

```python
#print response.text
query_text = response.text.replace(';',',')
#query_text = response.text
print query_text
```
